```python
import torch
import torchvision
import torchvision.transforms as transforms
```

```python
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,0.5,0.5), (0.5,0.5,0.5))])

#학습데이터 로드
trainset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)

trainloader = torch.utils.data.DataLoader(trainset, batch_size=4, shuffle=True, num_workers=2)
#평가데이터 로드
testset = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)

testloader = torch.utils.data.DataLoader(trainset, batch_size=4, shuffle=False, num_workers=2)

classes = ('plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 이미지 불러오기

#이미지를 보여주는 함수
def imshow(img):
    img = img / 2 + 0.5 #비정규화
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1,2,0)))
    
# 랜덤으로 학습이미지를 얻는다.
dataiter = iter(trainloader)
images, labels = dataiter.next()

# 이미지를 보여준다
imshow(torchvision.utils.make_grid(images))
# label(클래스 이름)을 출력한다
print(' '.join('%5s' % classes[labels[j]] for j in range(4)))
```

```python
import torch.nn as nn
import torch.nn.functional as F

# Convolution 신경망 정의 - 3채널 이미지

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(3, 6, 5)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(6, 16, 5)
        self.fc1 = nn.Linear(16 * 5 * 5, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10)
    
    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 16 * 5 * 5)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
    
net = Net()
```

```python
import torch.optim as optim

# 손실함수와 optim 정의

criterion = nn.CrossEntropyLoss() # 손실함수로 cross entropy 함수 사용
optimizer = optim.SGD(net.parameters(), lr=0.001, momentum=0.9) # optim은 SGD 사용
```

```python
# 네트워크 학습

for epoch in range(2):
    
    running_loss = 0.0
    for i, data in enumerate (trainloader, 0):
        # 입력 데이터를 얻는다.
        inputs, labels = data
        
        # 파라미터 그라디언트를 0으로 초기화한다.
        optimizer.zero_grad()
        
        # forward + backward + optimize
        outputs = net(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
        if i % 2000 == 1999:
            print('[%d, %5d] loss: %.3f' % (epoch + 1, i + 1, running_loss / 2000))
            runing_loss = 0.0
    
print('Finished Training')
```
